```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales_data = pd.DataFrame ({
    
'Order_ID' : [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010],
'Date' : ['2024-01-03', '2024-01-04', '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10', '2024-01-11', '2024-01-12'],
'Product' : ['Laptop', 'Headphones', 'Office Chair', 'T-Shirt', 'Smartphone', 'Desk', 'Jacket', 'Tablet', 'Sneakers', 'Bookshelf'],
'Category' : ['Electronics','Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Clothing', 'Furniture'],
'Quantity' : [1, 2, 1, 5, 1, 1, 2, 3, 2, 1],
'Unit_Price' : [1200, 150, 220, 25, 800, 450, 90, 300, 110, 180],
'Region' : ['Ontario', 'Ontario', 'Alberta', 'British Columbia', 'Ontario', 'Quebec', 'Alberta', 'British Columbia', 'Ontario', 'Quebec']

})
sales_data
```

```python
sales_data['Total_Sales'] = sales_data['Quantity'] * sales_data['Unit_Price'] 
sales_data
```

```python
sales_data["Date"] = pd.to_datetime(sales_data["Date"])
# sales_data.info()
```

## Hightest Selling Catergory

```python
Category_sales = sales_data.groupby('Category')['Total_Sales'].sum().sort_values(ascending = False)
# Electoniics generates the highest total sales with 3200 in revenue
# Furniture follows with 850, while Clothing is the lowest at 525.
plt.figure()
Category_sales.plot(kind="bar")
plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

##### High-priced items (like laptops, smartphones, and tablets) drive most of the revenue even when sold in small quantities.

##### Electronics clearly outperform other categories despite fewer units sold.


```python
region_sales = sales_data.groupby('Region')['Total_Sales'].sum().sort_values(ascending = False)


plt.figure()
region_sales.plot(kind="bar")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
# Ontario is the top-performing region, generating 2,520 in total sales.
# British Columbia comes next with 1,025, while Quebec and Alberta trail behind.
```

##### Ontario dominates sales, mainly due to multiple high-value electronics purchases.

##### Clothing sells more units but contributes less revenue because the prices are lower.

# Executive Summary

Overall, electronics are the biggest revenue driver for the business, bringing in far more sales than furniture or clothing. Even though clothing items sell in higher quantities, their lower prices limit their overall contribution. Ontario stands out as the strongest region, largely due to high-value electronics purchases. British Columbia performs moderately well, while Alberta and Quebec generate smaller revenues. This suggests that focusing on premium electronics products can significantly boost sales.
